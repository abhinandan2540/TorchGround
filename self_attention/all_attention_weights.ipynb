computing attention weights for all input tokens

```python

import torch
inputs = torch.tensor([
    [0.43, 0.15, 0.89],  # your        (x^1)
    [0.55, 0.87, 0.66],  # journey     (x^2)
    [0.57, 0.85, 0.64],  # starts     (x^3)
    [0.22, 0.58, 0.33],  # with       (x^4)
    [0.77, 0.25, 0.10],  # one        (x^5)
    [0.05, 0.80, 0.55]  # step        (X^6)
])
```

```python
print(inputs)
```

*calculating attention scores*

```python
attn_scores=torch.empty(6,6)
for i,x_i in enumerate(inputs):
    for j, x_j in enumerate(inputs):
        attn_scores[i,j]=torch.dot(x_i, x_j)

```

```python
print(attn_scores)
```

```python
# instead of for loops, we can use matrix multiplication
attn_scores=inputs@inputs.T

print(attn_scores)

```

*attention weights*

```python
attn_weights=torch.softmax(attn_scores,dim=-1)
print(attn_weights)

# dim=-1 specify that all row sum = 1
```

```python
print(attn_weights.sum(dim=-1))
```

*context vector*

```python
# context vector = by multiplying input embeded tokens with corresponding attention weights

all_context_vec=attn_weights@inputs
print(all_context_vec)
```
